```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# match_df = pd.read_csv("major-league-soccer-dataset/matches.csv", thousands=',')
# event_df = pd.read_csv("major-league-soccer-dataset/events.csv")
# table_df = pd.read_csv("major-league-soccer-dataset/all_tables.csv")
player_df = pd.read_csv("major-league-soccer-dataset/all_players.csv")
# gk_df = pd.read_csv("major-league-soccer-dataset/all_goalkeepers.csv")

df = player_df
df.Year = df.Year.astype('int') 
```

```python
def df_hist(df, col):
    values, counts = np.unique(df[col], return_counts=True)
    plt.scatter(values, counts, label=col)
    plt.legend()
    plt.show()
```

```python
df_hist(df[ df['SOG']>0 ], 'SOG')
```

```python
df_hist(df, 'Season')
```

```python
for name in df.Player:
    temp_df = df[df['Player']==name]
    plt.plot(temp_df['Year'], temp_df['G/90min'])
    
plt.show()
```
